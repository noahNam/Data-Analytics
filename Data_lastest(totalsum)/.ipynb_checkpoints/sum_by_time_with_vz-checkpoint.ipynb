```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a_df=pd.read_excel('a.xlsx')
```

```python
b_df=pd.read_excel('b.xlsx')
```

```python
c_df=pd.read_excel('c.xlsx')
```

```python
a_df.shape
```

```python
b_df.shape
```

```python
c_df.shape
```

```python
criteria=b_df.shape[0]
```

```python
a_df=a_df[:criteria]
c_df=c_df[:criteria]
```

```python
df=a_df+b_df+c_df
```

```python
df.date=[str(i)[:10] for i in df.date]
```

```python
df.head()
```

## 연별 합계

```python
year_df=df.copy()
```

```python
year_df['year']=[i[:4] for i in df.date]
```

```python
year_df.groupby('year').sum().T.sum()
```

```python
year_df.groupby('year').sum().T.sum().plot.pie(y='year', figsize=(10, 8),autopct='%1.1f%%')
plt.show()
```

# 월별 합계

```python
df['month']=[i.split('-')[0]+i.split('-')[1] for i in df.date]
```

## 월별 15분 단위 합계

```python
df.groupby('month').sum()
```

## 연월별 총 합계

```python
df.groupby('month').sum().T.sum().plot.line(y='quantity', figsize=(14, 8))
plt.show()
```

![color_box](color_box_by_month.png)

## 월별 총 합계

```python
df['month_without_year']=[i.split('-')[1] for i in df.date]
```

```python
df.groupby('month_without_year').sum().T.sum()
```

```python
df.groupby('month_without_year').sum().T.sum().plot.line(figsize=(14, 8))
plt.show()
```

![month_without_year](month_without_year.png)

```python
df.groupby('month').sum().T.sum().to_excel('moth_sum.xlsx')
df.groupby('month_without_year').sum().T.sum().to_excel('month_without_year.xlsx')
```

# 시간별 합계

```python
df=df.drop(['month'],axis=1)
```

```python
times=[str(i)[:2] for i in df.columns[1:]]
```

```python
time_df=df.T
time_df=time_df.drop(['date'])
```

```python
time_df.head()
```

```python
time_df['times']=times
```

## 시간 별 총 합계

```python
time_df.groupby('times').sum()
```

```python
time_df.groupby('times').sum().to_excel('time_sum.xlsx')
```

```python
t_df=df.groupby('date').sum().T
```

```python
t_df['times']=times
```

!['time_sum'](time_sum.png)

## 일별, 시간별 합계

```python
t_df.groupby('times').sum().T
```

```python
t_df.groupby('times').sum().T.to_excel('daily_time_sum.xlsx')
```
