```python
import pandas as pd
```

```python
df=pd.read_excel('c.xlsx')
```

```python
df.head()
```

# 월별 합계

```python
df['month']=[i.split('-')[0]+i.split('-')[1] for i in df.date]
```

## 월별 15분 단위 합계

```python
df.groupby('month').sum()
```

## 월별 총 합계

```python
df.groupby('month').sum().T.sum()
```

# 일별 합계

## 일별 15분 단위 합계

```python
df.groupby('date').sum()
```

## 일별 총 합계

```python
df.groupby('date').sum().T.sum()
```

# 시간별 합계

```python
df=df.drop(['month'],axis=1)
```

```python
times=[str(i)[:2] for i in df.columns[1:]]
```

```python
time_df=df.T
time_df=time_df.drop(['date'])
```

```python
time_df.head()
```

```python
time_df['times']=times
```

## 시간 별 총 합계

```python
time_df.groupby('times').sum()
```
